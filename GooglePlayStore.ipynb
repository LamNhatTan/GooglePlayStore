**Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Loading Dataset**

```python
df = pd.read_csv("googleplaystore.csv")
```

```python
# Handling Missing Values
df.dropna(inplace=True) # Drop rows with missing values
# Handling Duplicates
df.drop_duplicates(inplace=True) # Remove duplicates values
```

**Check again ensure clean dataset**

```python
# Check missing values
df.info()
```

```python
# Check duplicates
df.duplicated().sum()
```

**Handling columns**

```python
# Edit the "Installs" column
df["Installs"] = df["Installs"].str.replace("+","") # Take off "+" sign
df["Installs"] = df["Installs"].str.replace(",","") # Remove "," sign
df["Installs"] = df["Installs"].astype(int) # Convert data type to int

# Convert data type "Reviews" column to int
df["Reviews"] = df["Reviews"].astype(int)

df.head()
```

```python
# Export file
df.to_csv("GooglePlayStoreClean.csv")
```

# Exploratory Data Analysis

```python
# Bar Plot
plt.figure(figsize=(14,6))
sns.countplot(data=df, x="Category")
plt.xticks(rotation=90)
plt.ylabel("Count")
plt.title("Distribution of App Categories")
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(14,6))
sns.scatterplot(data=df,x="Reviews",y="Rating")
plt.xlabel("Number of Reviews")
plt.title("Relationship between Reviews and Rating")
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(14,6))
sns.boxplot(data=df, x ="Category",y="Installs")
plt.xticks(rotation = 90)
plt.title("Distribution of Installs across Category")
plt.show()
```

# Hypothesis Testing

```python
# Splitting dataset into free and paid apps
free_apps = df[df["Type"] == "Free"]
paid_apps = df[df["Type"] == "Paid"]
```

**Hypothesis zero is significant difference in the average rating between free and paid apps**

```python
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(free_apps["Rating"],paid_apps["Rating"])
t_stat
```

```python
p_value
```

```python
# Checking p_value
if p_value < 0.05:
    print("Reject null hypothesis: There is a significant diffenrence in ratings.")
else:
    print("Fail to reject null hypothesis: There is no significant diffenrence in ratings.")
```

```python
# Line plot of Average Rating over Time
plt.figure(figsize=(10,6))
sns.lineplot(data=df,x="Last Updated", y="Rating")
plt.xticks(rotation=90)
plt.ylabel("Average Rating")
plt.title("Average Rating Trend over Time")
```

```python
# Convert "Last Updated" to datatime
df["Last Updated"] = pd.to_datetime(df["Last Updated"])

# Set "Last Updated" as the index
df.set_index("Last Updated",inplace=True)

# Resample "Rating" to Monthly frequency
df_monthly = df["Rating"].resample("M").mean()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=df_monthly)
plt.xlabel("Date of Last Update (Monthly)")
plt.xticks(rotation=90)
plt.ylabel("Average Rating")
plt.title("Average Rating Trend over Time (Monthly)")
plt.show()
```
